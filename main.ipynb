#### **Least/Most  Fuel Efficient Car Per Year Section 1**

```python
import pandas as pd
import numpy as np

# Read the dataset
auto_dataset = pd.read_csv("Auto.csv", na_values=['?'])

# Define the range of years you're interested in
years_1970_1974 = [70, 71, 72, 73, 74]

# Dictionaries to store the least and most efficient cars
least_efficient_cars_1970_1974 = {}
most_efficient_cars_1970_1974 = {}

# Process each year
for year in years_1970_1974:
    cars = auto_dataset.loc[auto_dataset['year'] == year, ["year", "name", "mpg"]]
    least_efficient_cars_1970_1974[year] = cars.loc[cars["mpg"].idxmin()]
    most_efficient_cars_1970_1974[year] = cars.loc[cars["mpg"].idxmax()]

# Print the results
for year in years_1970_1974:
    print(f'The least efficient car in {year + 1900} was {least_efficient_cars_1970_1974[year]["name"]} with mpg of {least_efficient_cars_1970_1974[year]["mpg"]}')
    print(f'The most efficient car in {year + 1900} was {most_efficient_cars_1970_1974[year]["name"]} with mpg of {most_efficient_cars_1970_1974[year]["mpg"]}')
    print("\n")
```

#### **Average Weight Of Each Year**

```python
auto_dataset = pd.read_csv("Auto.csv", na_values=['?'])

weight_years_1970_1982 = [70,71,72,73,74,75,76,77,78,79,80,81,82]

average_weight_of_each_year = {}

for year in weight_years_1970_1982:
    cars = auto_dataset.loc[auto_dataset["year"] == year, ["year", "name", "weight"]]
    average_weight_of_each_year[year] = cars["weight"].mean().round(2)

for year in weight_years_1970_1982:
    print(f'The average weight of the year {year + 1900} was {average_weight_of_each_year[year]} lbs')
    print("\n")

 
```

#### **Average Weight Of Each Year Graph**

```python
years = [year + 1900 for year in weight_years_1970_1982]

average_weights = [average_weight_of_each_year[year] for year in weight_years_1970_1982]

plt.figure(figsize=(12,10))
plt.plot(years, average_weights, marker= "o", linestyle="-", color="blue", label="pounds(lbs)")
plt.xlabel("years")
plt.ylabel("weight")
plt.title("Average Weight Of Each Year from 1970-1982")
plt.legend()
plt.grid(True)
plt.show()
```

#### **Average MPG Of Each Year**

```python
auto_dataset = pd.read_csv("Auto.csv", na_values=['?'])

mpg_years_1970_1982 = [70,71,72,73,74,75,76,77,78,79,80,81,82]

average_mpg_of_each_year = {}

for year in mpg_years_1970_1982:
    cars = auto_dataset.loc[auto_dataset["year"] == year, ["name", "year", "mpg"]]
    average_mpg_of_each_year[year] = cars["mpg"].mean().round(2)

for year in mpg_years_1970_1982:
    print(f'-The average mpg of the year {year + 1900} was {average_mpg_of_each_year[year]} mpg ')
    print('\n')
    
```

#### **Average MPG Of Each Year Graph**

```python
mpg_years = [year + 1900 for year in mpg_years_1970_1982]

average_mpg = [average_mpg_of_each_year[year] for year in mpg_years_1970_1982]

plt.figure(figsize=(12,10))
plt.plot(mpg_years, average_mpg, marker="*", linestyle="-", color="red", label="miles per gallon (mpg)")
plt.xlabel("years")
plt.ylabel("mpg")
plt.title("Average mpg Of Each Year From 1970-1982")
plt.legend()
plt.grid(True)
plt.show()
```

#### **Conclusion**

After analyzing the different graphs and calculations, we can conclude that **there is a correlation between a decrease in total weight and an increase in total fuel efficiency in commercial cars from 1970 to 1982.** 
From the first year to the last year, the average weight decreased from 3372.79 lbs to 2434.17 lbs, a difference of 938.62 lbs. The average mpg increased from 17.69 mpg to 32 mpg, an increase of 14.31 mpg. However, this doesn't completely prove that a lighter weights improve fuel efficiency as I have only used on variable.  
Moreover, the dataset ends in 1982, so we can't conclude wether this trend continued or there were changes. 

Section 1 proved to be the section with the heaviest vehicles. It also contained the most fuel deficient car, the hi 1200d with 9.0 mpg. Although it contained the years with the worst fuel efficiency averages, the tendency in the mpg graph is positive, which means that cars were getting better at managing fuel. Section 1 is thus proving the conclusion.

Section 2 continues that same trend as section 1. Tho cars haven't improved beyond the general averages, cars were showing to be more lighter and fuel efficient. The decline in the weight section is more linear as well as that the increase in mpg is slightly linear. What this section does it connect to section 3 and it's abrupt changes, were the real findings are found.

Section 3 is were we really prove our conclusion. This section contains the most efficient car, which was the Mazda glc of 1980 with an efficiency of 46.6 mpg.
