#### **Least/Most  Fuel Efficient Car Per Year Section 1**

```python
import pandas as pd
import numpy as np

# Read the dataset
auto_dataset = pd.read_csv("Auto.csv", na_values=['?'])

# Define the range of years you're interested in
years_1970_1974 = [70, 71, 72, 73, 74]

# Dictionaries to store the least and most efficient cars
least_efficient_cars_1970_1974 = {}
most_efficient_cars_1970_1974 = {}

# Process each year
for year in years_1970_1974:
    cars = auto_dataset.loc[auto_dataset['year'] == year, ["year", "name", "mpg"]]
    least_efficient_cars_1970_1974[year] = cars.loc[cars["mpg"].idxmin()]
    most_efficient_cars_1970_1974[year] = cars.loc[cars["mpg"].idxmax()]

# Print the results
for year in years_1970_1974:
    print(f'The least efficient car in {year + 1900} was {least_efficient_cars_1970_1974[year]["name"]} with mpg of {least_efficient_cars_1970_1974[year]["mpg"]}')
    print(f'The most efficient car in {year + 1900} was {most_efficient_cars_1970_1974[year]["name"]} with mpg of {most_efficient_cars_1970_1974[year]["mpg"]}')
    print("\n")
```

#### **Lightest/Heaviest Car Per Year in Section 1**

```python

lightest_cars_1970_1974 = {}
heaviest_cars_1970_1974 = {}

for year in years_1970_1974:
    cars = auto_dataset.loc[auto_dataset["year"] == year, ["year", "name", "weight"]]
    lightest_cars_1970_1974[year] = cars.loc[cars["weight"].idxmin()]
    heaviest_cars_1970_1974[year] = cars.loc[cars["weight"].idxmax()]

for year in years_1970_1974:
    
    print(f'The lightest car from {year + 1900} was the {lightest_cars_1970_1974[year]["name"]} with a weight of {lightest_cars_1970_1974[year]["weight"]} lbs')
    print(f'The heaviest car from {year + 1900} was the {heaviest_cars_1970_1974[year]["name"]} with a weight of {heaviest_cars_1970_1974[year]["weight"]} lbs')
    print("\n")
    
```

**Weight Analysis**

The most characteristic detail about section one is that we find the heaviest and lightest cars in all of the three sections. The heaviest car is the Pontiac Safari (sw) from 1971 with a weight of 5140 lbs while we also find the lightest car which was the Datsun 1200 with a weight of 1613 lbs also from 1971. Based on this we can conclude that 1971 was the year with the biggest range of weights. A great amount of cars are above the average weight (2970 lbs), which tells us that in this specific years car manufacturers had inclinations to mid-weight to heavy cars.

The years with the most peculiar behavior are 1972 and 1973. In the distribution of data from 1972 we see that it's divided into two main groups: one far below the average weight and one far above the average weight. In fact, the ones that are close to the mean are the outliers. We can infer that this year car manufacturers really went for polar directions. 
Meanwhile the distribution of the year 1973 is more diverse as the range is really packed and there's no significant gaps among the data distribution. We see a great variety of below weight average cars as well as in the above average weight cars.
